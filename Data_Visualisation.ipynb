## 1 - Matplotlib <a class="anchor" id="Matplotlib"></a>

`Matplotlib` est la librairie graphique de Python. Elle permet de visualiser des graphiques de manière interactive dans un notebook Jupyter.

`pyplot` est l'interface de Matplotlib.

Pour pouvoir rediriger la sortie graphique de Matplotlib dans le notebook il faut utiliser le "Line magics" suivante:

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
#simple plot
plt.plot(range(11),'o')
plt.show()
```

A partir de l'interface `plt`, on peut controller les paramètres de la figure, tels que les légendes des axes.

```python
import random
plt.plot(random.sample(range(11),10), 'o')

# Axes
plt.ylabel("Random values")
plt.xlabel("Random samples")

plt.show()
```

Il existe différent types de plots:

```python
plt.figure(1)
plt.subplot(221)
plt.plot(random.sample(range(11),10), 'o')

plt.subplot(222)
plt.bar(random.sample(range(11),10),range(10))

plt.subplot(223)
plt.hist(random.sample(xrange(50),20 ), 10)

plt.subplot(224)
plt.barh(random.sample(range(11),10),range(10))

plt.show()
```

Les graphiques sont facilement supperposables.

```python
plt.plot(random.sample(range(11),10), 'o')
plt.plot(random.sample(range(11),10), 'r')
plt.show()
```
